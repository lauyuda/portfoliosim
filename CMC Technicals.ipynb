```python
import pandas as pd
```

```python
df = pd.read_csv('data/CMC200.csv', parse_dates=['Date'], index_col='Date')
```

```python
df
```

```python
df.plot()
```

```python
# drop the missing data; disconnection in the graph above
df = df.dropna()
```

```python
df.iloc[400:]
```

```python
df['sma200'] = df['Close'].rolling(window=200).mean()
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=[16,8])
plt.plot(df['Close'],label='Closing Price')
plt.plot(df['sma200'],label='SMA 200 Days')
plt.title('200 SMA')

plt.show()
```

```python
df.tail()
```

```python
df['30 Day MA'] = df['Adj Close'].rolling(window=20).mean()
df['30 Day STD'] = df['Adj Close'].rolling(window=20).std()
df['Upper Band'] = df['30 Day MA'] + (df['30 Day STD'] * 2)
df['Lower Band'] = df['30 Day MA'] - (df['30 Day STD'] * 2)
```

```python
# Simple 30 Day Bollinger Band for Crypto (2019-2021)
df[['Adj Close', '30 Day MA', 'Upper Band', 'Lower Band']].plot(figsize=(16,8))
plt.title('30 Day Bollinger Band for Crypto')
plt.ylabel('Price (USD)')
plt.show()

# set style, empty figure and axes
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(111)

# Get index values for the X axis for facebook DataFrame
x_axis = df.index.get_level_values(0)

# Plot shaded 21 Day Bollinger Band for Facebook
ax.fill_between(x_axis, df['Upper Band'], df['Lower Band'], color='grey')

# Plot Adjust Closing Price and Moving Averages
ax.plot(x_axis, df['Adj Close'], color='blue', lw=2)
ax.plot(x_axis, df['30 Day MA'], color='black', lw=2)

# Set Title & Show the Image
ax.set_title('30 Day Bollinger Band For CMC200')
ax.set_xlabel('Date (Year/Month)')
ax.set_ylabel('Price(USD)')
ax.legend()
plt.show()
```
